# Data Understanding

```python
data.head()
```

```python
data= data[data['Order Date'].str[0:2] !='Or']
```

```python
data['Month'] = data['Order Date'].str[0:2]
data['Month'].astype("int32")
```

```python
data.head()
```

```python
data.info()
```

```python
months_name = {'01':'jan', '02': 'feb', '03': 'mar','04':'apr', '05':'may', '06':'jun', '07':'july',
              '08':'aug', '09':'sep', '10':'oct', '11':'nov', '12': 'dec'}

data['Month'] = data['Month'].apply(lambda x: months_name[x])
data.head()
```

```python
data.columns
```

```python
for i in ['Quantity Ordered', 'Price Each']:
    data[i]= data[i].astype('float')
   
```

```python
data.info()
```

```python
data['Sales'] = data['Quantity Ordered'] * data['Price Each']
data.head()
```

```python
highest_sales = data.groupby('Month').sum()
highest_sales.head()
```

```python
plt.bar(highest_sales.index,highest_sales['Sales'])
```

# which city sold most of the productes

```python
data['city'] = data['Purchase Address'].apply(lambda x:x.split(',')[1])
```

```python
data.head()
```

```python
city_sold= data.groupby("city").sum()
```

```python
plt.bar(city_sold.index, city_sold['Quantity Ordered'])
plt.xticks(rotation= 90)
plt.show()
```

```python
data['Order ID'].duplicated().sum()
```

```python
data.shape
```

```python
duplicate_data = data[data['Order ID'].duplicated(keep = False)]
```

```python
duplicate_data
```

```python
duplicate_data['product_bundle'] = duplicate_data.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
```

```python
duplicate_data.head()
```

```python
duplicate_data=duplicate_data[['Order ID','product_bundle']].drop_duplicates()
duplicate_data.head()
```

```python
from itertools import combinations
from collections import Counter
```

```python
count = Counter()
for row in duplicate_data['product_bundle']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 3)))
count.most_common(10)
```
