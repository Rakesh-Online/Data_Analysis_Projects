```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('zomato.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
df.describe()
```

```python
# finding sum of null values
df.isnull().sum()
```

```python
# List comprehension is an other way to get the null valued column 
[features for features in df.columns if df[features].isnull().sum() > 0] 
```

```python
df_country = pd.read_excel('Country-Code.xlsx')
```

```python
df_country.head()
```

```python
# we can see country code below so we have to merge the two two datasets
df.columns
```

```python
# we can use merge to do SQL type of joins 
final_df = pd.merge(df,df_country,on= 'Country Code', how = 'left')
```

```python
final_df.head()
```

```python
country_name = final_df.Country.value_counts().index
```

```python
country_values =final_df.Country.value_counts().values
```

```python
# below are top 3 countries
plt.pie(country_values[:3], labels = country_name[:3], autopct = '%1.2f%%')
plt.show()
```

##  **Obeservation1:** 
- Zomato maximum records or transactions are from india after that USA and then United kingdom 

```python
final_df.columns
```

```python
rating = final_df.groupby(['Aggregate rating', 'Rating color','Rating text']).size().reset_index().rename(columns ={0 :'Rating Count'})
rating
```

## Obesarvation 2: 
1. When rating is between 4.5 to 4.9 --> excellent
2. When rating is between 4.0 to 4.4 --> very good
3. when rating is between 3.5 to 3.9 ---> good
4. When rating is between 3.0 to 3.4 --> average
5. When rating is between 2.5 to 2.9 --> average
6. When rating is between 2.0 to 2.4 --> poor

```python
# we can use the palette for the colors we want as per the requirement for now we are not using it 
plt.figure(figsize =(10,8))
sns.barplot(x='Aggregate rating', y= 'Rating Count',data = rating)
```

## observation-3 : 
1. Not rated count is very high
2. Maximum number of ratingare between 2.5 to 3.4

```python
## count plot
sns.countplot(x= 'Rating color', data = rating)
```
